```python
import plotly.graph_objects as go
import plotly.offline as pyo  # Saving HTML

# Function to plot world map using CO2 emissions
def plot_world_map_co2():
    fig = go.Figure()
    
    # Loop through the years 2000-2021 for animation, excluding 2020
    years = filtered_temps_emissions_table['Year'].unique()
    years = [year for year in years if year != 2020]  # Exclude 2020
    
    # Filter the U.S. data separately
    us_data = filtered_temps_emissions_table[filtered_temps_emissions_table['Country'] == 'United States']
    
    for year in years:
        # Filter the data for the current year
        filtered_df = filtered_temps_emissions_table[filtered_temps_emissions_table['Year'] == year]

        # Set a smaller range for zmin and zmax to exaggerate differences
        zmin_value = filtered_temps_emissions_table['Value_co2_emissions_kt_by_country'].quantile(0.1)  # 10th percentile
        zmax_value = filtered_temps_emissions_table['Value_co2_emissions_kt_by_country'].quantile(0.9)  # 90th percentile

        # Create a choropleth trace for the current year
        trace = go.Choropleth(
            locations=filtered_df['Country'],  # Country names
            z=filtered_df['Value_co2_emissions_kt_by_country'],  # CO2 emissions for color scaling
            locationmode='country names',
            colorscale=[  # Custom colorscale with more contrast
                [0, "rgb(255, 255, 178)"],  # Light yellow for lower emissions
                [0.25, "rgb(254, 204, 92)"],
                [0.5, "rgb(253, 141, 60)"],
                [0.75, "rgb(240, 59, 32)"],
                [1, "rgb(189, 0, 38)"]  # Dark red for higher emissions
            ],
            colorbar=dict(title='CO2 Emissions (kt)'),
            zmin=zmin_value,  # Narrow the range
            zmax=zmax_value,  # Narrow the range
            visible=False,  # Set all traces to invisible
            hoverinfo='location+z'  # Only show country name and CO2 emissions in the hover popup
        )
        
        # Add the trace to the figure
        fig.add_trace(trace)

    # Set the first trace to be visible
    fig.data[0].visible = True
    
    # Create animation steps for the slider
    steps = []
    for i in range(len(years)):  # Update for all traces
        # Get U.S. emissions for the current year
        us_year_data = us_data[us_data['Year'] == years[i]]
        us_emissions = us_year_data['Value_co2_emissions_kt_by_country'].values[0] if not us_year_data.empty else 0
        
        step = dict(
            method='update',
            args=[{'visible': [False] * len(fig.data)},  # Hide all traces
                  {'title_text': f'CO2 Emissions Map - {years[i]}'}],  # Set title with the year
            label=str(years[i]),  # Label for the slider step
        )
        step['args'][0]['visible'][i] = True  # Make current year world trace visible
        
        # Dynamically update the annotation for U.S. emissions
        step['args'][1]['annotations'] = [dict(
            text=f"United States CO2 Emissions ({years[i]}): {us_emissions:.2f} kt",
            x=0.5,  # Horizontal center
            y=-0.05,  # Adjusted to avoid slider overlap
            xref="paper",  # Position relative to the entire figure
            yref="paper",
            showarrow=False,  # No arrow
            font=dict(size=12, color="blue"),  # Style the text
            xanchor="center",  # Center the text horizontally
            yanchor="top"  # Align the text to the top of the annotation position
        )]
        steps.append(step)
    
    # Create the slider
    sliders = [dict(
        active=0,
        steps=steps,
        currentvalue={"prefix": "Year: "}
    )]
    
    # Update layout to include the slider and adjust other properties
    fig.update_layout(
        title_text='CO2 Emissions by Country (2000-2019) - Highlighting U.S.',
        geo=dict(
            showframe=True,
            showcoastlines=True,
            projection_type='natural earth'
        ),
        sliders=sliders,
        height=600,
        width=1000,
        margin=dict(l=50, r=50, t=50, b=120)  # Increased bottom margin for space
    )
    
    # Save the figure to an HTML file
    pyo.plot(fig, filename='co2_emissions_world_map_with_us.html')

# Call the function to display the world map
plot_world_map_co2()
```
